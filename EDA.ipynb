```python
def get_sign(x):
    if x > 0:
        return 'Positive'
    elif x < 0:
        return 'Negative'
    else:
        return 'Zero'
```

```python

df['sign'] = df['Excess Return'].apply(lambda x: get_sign(x))
df_1['sign'] = df_1['Excess Return'].apply(lambda x: get_sign(x))
df_2['sign'] = df_2['Excess Return'].apply(lambda x: get_sign(x))
df_3['sign'] = df_3['Excess Return'].apply(lambda x: get_sign(x))
df_4['sign'] = df_4['Excess Return'].apply(lambda x: get_sign(x))

# Display the resulting DataFrame
df.groupby('sign')[['ID']].count().plot(kind = 'bar')
```

```python
fig, ax = plt.subplots(2,2, figsize=(16,12))


df_1.groupby('sign')[['ID']].count().plot(ax=ax[0,0],kind  ='bar')
ax[0,0].set_title('Return against Benchmark 1990 - 1995')
ax[0,0].set_ylabel("Count")

df_2.groupby('sign')[['ID']].count().plot(ax=ax[0,1], kind='bar')
ax[0,1].set_title('Return against Benchmark 1995 - 2000')
ax[0,1].set_ylabel("Count")

df_3.groupby('sign')[['ID']].count().plot(ax=ax[1,0], kind='bar')
ax[1,0].set_title('Return against Benchmark 2000 - 2005')
ax[1,0].set_ylabel("Count")

df_4.groupby('sign')[['ID']].count().plot(ax=ax[1,1], kind='bar')
ax[1,1].set_title('Return against Benchmark 2005 - 2010')
ax[1,1].set_ylabel("Count")
```
